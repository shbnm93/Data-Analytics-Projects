# Visualize the Distribution of Employment Status:

Create a bar plot to visualize the distribution of participants based on their employment status.

```python
# Visualize the distribution of employment status
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='calisma_durumu')
plt.title("Distribution of Employment Status")
plt.xticks(rotation=45)
plt.show()
```

# Explore the Relationship Between Employment Status and Education Level:

Investigate the relationship between 'calisma_durumu' and 'egitim_durumu'.

```python
# Create a cross-tabulation and visualize the relationship
cross_tab_education_employment = pd.crosstab(df['egitim_durumu'], df['calisma_durumu'])
cross_tab_education_employment.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title("Relationship Between Education Level and Employment Status")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title='Employment Status')
plt.show()
```

# Visualizing Program Distribution:

```python
# Choose the top N programs to display
top_n_programs = 10  

# Filter the dataframe for the top N programs
top_programs_df = df['program'].value_counts().nlargest(top_n_programs).index
filtered_df = df[df['program'].isin(top_programs_df)]
```

```python
# Visualize the distribution of top N programs
plt.figure(figsize=(15, 8))
sns.countplot(data=filtered_df, x='program', order=filtered_df['program'].value_counts().index)
plt.title(f"Top {top_n_programs} Programs - Distribution of Participants")
plt.xlabel("Program")
plt.ylabel("Count")
plt.xticks(rotation=75)
plt.show()
```

```python
# Visualize the distribution of educational levels
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='egitim_durumu', order=df['egitim_durumu'].value_counts().index)
plt.title("Distribution of Participants Across Educational Levels")
plt.xlabel("Educational Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
